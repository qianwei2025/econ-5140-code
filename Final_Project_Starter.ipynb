## Option 7 — Causal: Regression discontinuity (elections, schools, or birth outcomes)

**Question (general):** Use a **regression discontinuity**: compare units **just above** vs **just below** a cutoff in a **running variable**, so the estimand is **local** to the threshold—not a global average effect. State your **causal question** explicitly (see each data block below).

**Why it fits causal inference:** **Sharp** or **fuzzy** RD uses variation from a **rule** near a threshold—core material from **regression discontinuity** (local comparability, manipulation checks). This option **does not** reuse the **synthetic control / Prop 99** exercise from lecture.

**Data (pick one)—each row = one causal question + where to get the data:**

| Data source | What causal question you should answer (target estimand) |
|-------------|------------------------------------------------------------|
| **Lee 2008** | Effect of **winning vs losing** a *very close* House race at **margin = 0** on **next-election vote share** (local **incumbency** contrast). |
| **Pischke / `graderep`** | Effect of **grade retention vs promotion** at the **rule’s cutoff** on a **later outcome**—**local** to the threshold, not for all retained students. |
| **Kaggle birth weight** | Effect of crossing a **birth-weight cutoff** (e.g. **1500 g** VLBW) on **health outcomes**, comparing babies **just above** vs **just below**; address **fuzzy** RD and **heaping**. |

- **U.S. House elections (Lee 2008)**  
  - **Causal question:** What is the effect of **winning vs losing** a very close U.S. House race (as good as randomly assigned at the **0** margin cutoff) on **electoral performance in the next election** (e.g. vote share)? Intuition: **incumbency** vs **open-seat** dynamics for candidates who barely cross the threshold.  
  - **Data:** Official replication **`Lee2008.zip`** — [MHE data archive](https://economics.mit.edu/people/faculty/josh-angrist/mhe-data-archive) (section **Lee (2008)**) or [direct download](https://economics.mit.edu/sites/default/files/inline-files/Lee2008%20%281%29.zip). Read with `pandas.read_stata` (or CSV if you export); running variable = **Democratic margin** in the prior election; outcome = **vote share** in the next election — see Lee (2008), *Journal of Econometrics*.

- **Grade repetition / school rules (Pischke 2007; MHE replication)**  
  - **Causal question:** What is the effect of **grade retention** (being held back) vs **promotion** for students **near the cutoff** of a **school-system rule** (e.g. test-score threshold) on a **subsequent outcome** (e.g. later achievement or attainment)? The estimand is the **local** effect of retention at the margin, not for all retained students.  
  - **Data:** **`graderep.dta`** — same [MHE archive](https://economics.mit.edu/people/faculty/josh-angrist/mhe-data-archive) ([`.dta`](https://economics.mit.edu/sites/default/files/inline-files/graderep.dta), [`.do`](https://economics.mit.edu/sites/default/files/inline-files/graderep.do)); align variable names and cutoff with the do-file / MHE discussion.


**Methods (examples):** Binned or local **scatterplots** on either side of the cutoff; **local linear** RD (simple implementation with polynomials or bins, or a dedicated package such as **`rdrobust`** if you use it); **two bandwidths** or a **bandwidth robustness** table; optional discussion of **density** / manipulation at the cutoff.

## Option 10 — Road safety & accident severity

**Question:** How are **visibility**, **weather**, and **road context** related to **accident severity** after conditioning on **location** and other observables?

**Is this “causal”?** In the **raw** file, weather and lighting are **not** randomly assigned. A regression of severity on rain/night is usually a **conditional association** (risk decomposition / **prediction** under **selection on observables**), **not** a clean causal effect of weather on crashes. That is still a valid **econometrics** project if you are **explicit**:
- **Default (recommended):** Frame results as **associational** or **partially adjusted** comparisons; discuss **omitted variables** (driver behavior, traffic volume, reporting).
- **Stronger causal flavor (only if you earn it):** Add a **design** from outside the CSV—e.g. merge a **policy** with clear **timing** (state law change) and use **DID** / **event study** on **aggregated** outcomes; or a **credible** **RD** (rare here); or an **IV** you can defend. **Do not** call OLS/logit on weather a “causal effect” without that extra structure.

**Data:** [US Accidents (2016–2023)](https://www.kaggle.com/datasets/sobhanmoosavi/us-accidents) — **very large**; use a **random subsample** and/or **one state**. Read column definitions on the dataset page.

**Methods (examples):** Ordered logit or linear **severity**; **logistic** for serious injury; **rich controls** or **DML** for high-dimensional $X$; **state/month FE** on **aggregates** if you build a panel; **sensitivity** to specification.
