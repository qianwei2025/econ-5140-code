## Option 10 — Road safety & accident severity (Kaggle — US Accidents; read the “causal” scope carefully)

**Question:** How are **visibility**, **weather**, and **road context** related to **accident severity** after conditioning on **location** and other observables?

**Is this “causal”?** In the **raw** file, weather and lighting are **not** randomly assigned. A regression of severity on rain/night is usually a **conditional association** (risk decomposition / **prediction** under **selection on observables**), **not** a clean causal effect of weather on crashes. That is still a valid **econometrics** project if you are **explicit**:
- **Default (recommended):** Frame results as **associational** or **partially adjusted** comparisons; discuss **omitted variables** (driver behavior, traffic volume, reporting).
- **Stronger causal flavor (only if you earn it):** Add a **design** from outside the CSV—e.g. merge a **policy** with clear **timing** (state law change) and use **DID** / **event study** on **aggregated** outcomes; or a **credible** **RD** (rare here); or an **IV** you can defend. **Do not** call OLS/logit on weather a “causal effect” without that extra structure.

**Data:** [US Accidents (2016–2023)](https://www.kaggle.com/datasets/sobhanmoosavi/us-accidents) — **very large**; use a **random subsample** and/or **one state**. Read column definitions on the dataset page.

**Methods (examples):** Ordered logit or linear **severity**; **logistic** for serious injury; **rich controls** or **DML** for high-dimensional $X$; **state/month FE** on **aggregates** if you build a panel; **sensitivity** to specification.

**Minimum:** Clear statement whether your estimates are **associational** or **identified**; sampling plan; limitations on **reporting** and **selection** into the dataset.
